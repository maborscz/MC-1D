```python
import numpy as np

import time

import matplotlib.pyplot as plt
```

```python
N = 5000000

u = np.random.uniform(0, 1, N)

start = time.time()
u_sort1 = np.sort(u)
print(time.time() - start)

start = time.time()
cumsum = u.cumsum()
u_sort2 = cumsum / cumsum[-1]
print(time.time() - start)

plt.hist(u_sort1, bins=100, density=True, alpha=0.5, label='sort')
plt.hist(u_sort2, bins=100, density=True, alpha=0.5, label='cumsum')
plt.ylim(0.9, 1.1)
plt.legend()
plt.show()
```
